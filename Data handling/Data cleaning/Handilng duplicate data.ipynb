# Data cleaning

### Handling duplicate data

Duplicate observations most frequently arise during data collection, such as when we:

*   Combine datasets from multiple places
*   Scrape data ( Collect Data through web scraping)
* Receive data from clients/other departments

if we're using a web scraper we may happen to scrape the same webpage more than once, or the same information from two different pages. Whatever the reason, duplication can lead us to make incorrect conclusions by leading us to believe that some observations are more common than they really are.

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name" : ["John", "Rock", "Nail", "Tom", "John", "John"], "Rank" : [1, 2, 3, 6, 1, 1], "Marks" : [98, 59, 97, 99, 98, 98]})
df
```

```python
df.duplicated(keep=False) # Duplicates are show as "True"
```

```python
df.duplicated(keep=False).sum()
```

```python
df.duplicated().sum() #Here keep default value is First, ignore ther first occurrence row it set as valid & others (more than 1) as duplicates
```

```python
df.loc[df.duplicated(keep=False)]   # Print the all duplicated rows
```

```python
# keep : {"first", "last", "False"}, default "First"
# Drop duplicates except the first occurrence , one "John" row is there
df1 = df.drop_duplicates(keep="first")
df1
```

```python
df2 = df.drop_duplicates(keep=False) # will delete every duplicates rows( All "John" rows)
df2
```

```python
# drop duplicates except the last occurrence
df3 = df.drop_duplicates(keep="last")
df3
```

```python
df
```

```python
df.drop_duplicates(keep="first") # inplace = False, not update original one
```

```python
df
```

```python
df.drop_duplicates(keep="first", inplace=True) # inplace = True, update original one
df
```
