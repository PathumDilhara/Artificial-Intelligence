### Group By technique

```python
df = pd.DataFrame({"Name": ["sam", "kane", "Jane", "Sammie", "David"],
                  "Age": [23, 21, 25, 25, 27],
                  "Gender": ["M", "M", "F", "F", "M"],
                  "Marks": [78, 65, 78, 89, 76]})
df
```

```python
data_online_retail_germany.groupby(["InvoiceNo", "Description"])["Quantity"].sum() # let group with two arrgument to get sum of quantity
```

```python
data_online_retail_germany.groupby(["InvoiceNo", "Description"])["Quantity"].sum().unstack() # above one is called stack, convert it into unstak
```

```python
basket = data_online_retail_germany.groupby(["InvoiceNo", "Description"])["Quantity"].sum().unstack().fillna(0)
basket # all NaN will be replaced with 0
```

```python
basket = basket.astype(int) # float ---> int
basket
```

```python
def encode_units(x):
    if x>=1:
        return 1
    else:
        return 0
```

```python
basket = basket.applymap(encode_units)
```

```python
apr_result= apriori(basket, min_support=0.05, use_colnames=True)
```

```python
apr_result
```

```python
rules = association_rules(apr_result, metric="lift", min_threshold=1) # lift >= 1 ---> good
rules
```

```python
rules[(rules["lift"]>3) & (rules["confidence"]>0.5)]
```
